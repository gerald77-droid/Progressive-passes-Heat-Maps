```python
# Example criteria:
# Minimum pass length for a pass to be considered progressive
min_pass_length = 15  # Adjust as needed
progressive_passes=pass_events[(pass_events['pass_length']>=min_pass_length)]
num_progressive_passes = len(progressive_passes)
print("Number of progressive passes:", num_progressive_passes)
```

```python
import numpy as np

# Drop rows with NaN values in the 'location' column
df = df.dropna(subset=['location'])

# Convert float types to strings
df['location'] = df['location'].astype(str)
df['pass_end_location'] = df['pass_end_location'].astype(str)

# Extract x and y coordinates from pass_location and pass_end_location
df['pass_start_x'] = df['location'].apply(lambda loc: eval(loc)[0] if loc != 'nan' else None)
df['pass_start_y'] = df['location'].apply(lambda loc: eval(loc)[1] if loc != 'nan' else None)
df['pass_end_x'] = df['pass_end_location'].apply(lambda loc: eval(loc)[0] if loc != 'nan' else None)
df['pass_end_y'] = df['pass_end_location'].apply(lambda loc: eval(loc)[1] if loc != 'nan' else None)

pitch = Pitch(pitch_color="green", line_color="white", pitch_type='statsbomb')
fg, ax = pitch.draw()

for index, row in df.iterrows():
    start_x, start_y = row['pass_start_x'], row['pass_start_y']
    end_x, end_y = row['pass_end_x'], row['pass_end_y']
    if start_x is not None and start_y is not None and end_x is not None and end_y is not None:
        ax.plot([start_x, end_x], [start_y, end_y], color='blue', linewidth=1)

plt.title('Nigeria AFCON Final Passes')
plt.show()
```

```python
selected_columns=['team', 'player','location','period','duration','minute','play_pattern','timestamp','possession']
# Select specific columns

events_selected = events[selected_columns]

# Filter for the Nigerian team and reset index
events_filtered = events_selected[events_selected['team'] == 'Nigeria'].reset_index(drop=True)

```

```python
events_filtered.columns
```

```python
players_location=events_filtered['location'].to_list()
```

```python
# Filter out non-iterable elements from players_location
player_positions = [pos for pos in players_location if isinstance(pos, (list, tuple))]

# Extract x and y positions from filtered player_positions
x_positions = [pos[0] for pos in player_positions]
y_positions = [pos[1] for pos in player_positions]

plt.figure(figsize=(10, 7))
plt.hexbin(x_positions, y_positions, gridsize=20, cmap='magma')
plt.colorbar(label='Number of Events')
plt.title('Heat Map for {}'.format('Ademola lookman'))
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.gca().invert_yaxis()  # Invert y-axis to match soccer field orientation
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch

# Assuming you have loaded your data into a DataFrame called 'statsbomb_data'
# and filtered it to include only the events of the desired player
player_positions = [pos for pos in players_location if isinstance(pos, (list, tuple))]

x_positions = [loc[0] for loc in player_positions]
y_positions = [loc[1] for loc in player_positions]

# Create a pitch object
pitch = Pitch(pitch_type='statsbomb', pitch_color='green')

# Set up the figure
fig, ax = plt.subplots(figsize=(10, 7))

# Calculate density using numpy's histogram2d
density, x_edges, y_edges = np.histogram2d(x_positions, y_positions, bins=50)

# Plot the heatmap using imshow
sc = ax.imshow(density.T, extent=[0, 120, 0, 80], cmap='magma', alpha=0.8, origin='lower')

# Add colorbar
cbar = fig.colorbar(sc, ax=ax, label='Number of Events')

# Overlay the pitch on top
pitch.draw(ax=ax)
plt.gca().invert_yaxis()

# Add title and axis labels
plt.title('Heat Map')
plt.xlabel('X Position')
plt.ylabel('Y Position')

# Show the plot
plt.show()

```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch

# Assuming you have loaded your data into a DataFrame called 'statsbomb_data'
# and filtered it to include only the events of the desired player
player_positions = [pos for pos in players_location if isinstance(pos, (list, tuple))]

x_positions = [loc[0] for loc in player_positions]
y_positions = [loc[1] for loc in player_positions]

# Create a pitch object
pitch = Pitch(pitch_type='statsbomb', pitch_color='green')

# Set up the figure
fig, ax = plt.subplots(figsize=(10, 7))

# Plot the KDE plot using Seaborn's kdeplot function
sns.kdeplot(x=x_positions, y=y_positions, shade=True,shade_lowest=False,alpha=.5,n_levels=10,cmap="magma", ax=ax)

# Overlay the pitch on top of the KDE plot
pitch.draw(ax=ax)
plt.gca().invert_yaxis()

# Add title and axis labels
plt.title('Ademola Lookman Heat Map')
plt.xlabel('X Position')
plt.ylabel('Y Position')

plt.xlim(0,120)
plt.ylim(0,80)

# 
 # Invert y-axis to match soccer field orientation
plt.show()
```
